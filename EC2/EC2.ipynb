## EC2 - Elastic Compute Cloud

* Key pair is essential in Ec2 instances. It is used to connect remote instances to local machine using SSH.

* In advance details --> User Data<br>
    we can pass some scripts, or some comments to our EC2 instance to execute at the first launch of our EC2 instance and i will exectute only once.
    Bootstraping means automate tasks at boot time of instance. set of instructions we want to run every time our instance starts.
    
* Every time we start an instance, its pulic id changes.

* EC2 instances can have many security groups attached to it.

* And one security group can be attached to more than 1 EC2 instance.

* command to connect local machine to EC2 instance using SSH.
```ssh ec2-user@public-id-address```

    Amamzon Linux 2 AMI has one user already set up for us and That user is names as ec2-user.
    
* Steps to connect local machine to remote instance.

    1. move to certificate directory.
    
    2. chmod 0400 certificate.pem(0400 allows owner to read the file)
    
    3. ssh -i certificate.pem ec2-user@public-ip-address
    
* Always use IAM roles policies for ec2 instances.

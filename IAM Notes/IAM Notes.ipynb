## IAM Notes



* User and groups are created to give them permission so that they can use AWS services.

* In AWS we dont allow any person to do every thing, this means no one has permission to do every thing except root user.

* In AWS we apply LEAST PRIVILEDGE PRINCIPLE(LPP), which means dont give more permission than user needs.

* Tags are used to provide extra information and track, organise or control access for users.

* If we attch a policy at Group level, then all the users in that group will inherit that policy. That policy will be applied to all tye users in that group. 

* If a user doesn't belong to any group then we can attch to them directly to specific policy or we can use inline policy for that specific user.

* If we remove a user from specific group, then all the policies permissions, and access of that group will be revoked from that user.

* Access Keys are helpful if we use AWS CLI or AWS SDK.

* To create access keys we never use root account, we can use IAM user account or any other user account, who is going to use that AWS CLI or AWS SDK.

* IAM Roles --> Some services needs to perform some actions on our behalf, on our account. to do so, we assign permissions to AWS services with IAM Roles.

* IAM Roles are just like Users, but they are intended to be used not by physical user, but they will be used by AWS services.

* Roles allows entities in AWS to get credentials for short duration and to do whatever they need to do.

* AWS is responsible for Infrastructures and users are responsible for how to use that infrestructure.

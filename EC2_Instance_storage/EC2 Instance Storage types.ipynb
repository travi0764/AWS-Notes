# EC2 Instance Storage

<b> Now we will look at different storage options for EC2 Instances.</b>

* Elastic Block Store(EBS) :- allows our instances to persist data even after their termination, this means we can recreate an instance and mount to the same EBS volume from before, and we will get back our data.

* Note : - CCP(Certified Cloud Practitioner allows EBS to be mounted to one only one instance(one to one mapping). Associate level(Solution, Architect, SysOps) allows "multi-attach" feature for some EBS.

* When we create an EBS Volume, There is an option to tick (Delete on termination). By default it is ticked for root volume and not ticked for EBS volume.
